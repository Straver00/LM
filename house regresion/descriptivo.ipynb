```python
import pandas as pd  #  type: ignore
import matplotlib.pyplot as plt  #  type: ignore

df = pd.read_csv("../files/input/house_data.csv")

features = df[
    [
        "bedrooms",
        "bathrooms",
        "sqft_living",
        "sqft_lot",
        "floors",
        "waterfront",
        "condition",
    ]
]

target = df[["price"]]
```

```python
features.bathrooms.value_counts().sort_index().plot(kind="bar")
```

```python
features.sqft_living.plot(kind="hist", edgecolor="white", bins=20)
```

```python
features.sqft_lot.plot(kind="hist", edgecolor="white", bins=10)
```

```python
features.floors.value_counts().sort_index().plot(kind="bar")
```

```python
features.condition.value_counts().sort_index().plot(kind="bar")
```

```python
# Supongamos que ya tienes el objeto `target` definido
target.plot(kind="hist", edgecolor="white", bins=20)

# Formatear los valores del eje y en decenas de miles
plt.gca().get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x / 10000)}0k'))

plt.show()
```
